```python
print("=== SOLUSI PRAKTIKUM SISTEM I/O (MINGGU 12) ===\n")

# ==============================================================================
# SOAL 6: Interrupt Overhead (C3) [cite: 517-526]
# ==============================================================================
def solve_soal_6():
    print("--- SOAL 6: Interrupt Overhead ---")
    cpu_speed_ms = 2000      # instruksi/ms
    cpu_speed_sec = cpu_speed_ms * 1000 # konversi ke detik
    overhead_per_int = 100   # instruksi
    int_rate = 200           # interrupt/detik

    # a. Total instruksi untuk interrupt
    total_int_instr = int_rate * overhead_per_int

    # b. Persentase CPU Overhead
    percent_overhead = (total_int_instr / cpu_speed_sec) * 100

    # c. Sisa CPU untuk user
    user_percent = 100 - percent_overhead

    print(f"a. Total Instruksi Interrupt: {total_int_instr} instruksi/detik")
    print(f"b. CPU Overhead: {percent_overhead}%")
    print(f"c. Sisa CPU User: {user_percent}%")

# ==============================================================================
# SOAL 7: Cache Performance (C3) [cite: 530-539]
# ==============================================================================
def solve_soal_7():
    print("\n--- SOAL 7: Cache Performance (AMAT) ---")
    hit_time = 5       # ns
    miss_penalty = 100 # ns

    # a. AMAT dengan Hit Rate 92%
    h1 = 0.92
    amat_a = hit_time + ((1 - h1) * miss_penalty)

    # b. AMAT dengan Hit Rate 98%
    h2 = 0.98
    amat_b = hit_time + ((1 - h2) * miss_penalty)

    # c. Peningkatan Performa
    improvement = ((amat_a - amat_b) / amat_a) * 100

    print(f"a. AMAT (92%): {amat_a:.2f} ns")
    print(f"b. AMAT (98%): {amat_b:.2f} ns")
    print(f"c. Improvement: {improvement:.2f}%")

# ==============================================================================
# SOAL 8: DMA vs Interrupt (C3) [cite: 544-555]
# ==============================================================================
def solve_soal_8():
    print("\n--- SOAL 8: DMA vs Interrupt ---")
    file_size_mb = 8
    file_size_bytes = 8 * 1024 * 1024
    block_size = 512 # bytes
    int_overhead = 20 # microsecond (us) per interrupt
    dma_overhead_total = 20 # us total

    # a. Jumlah Block
    num_blocks = file_size_bytes / block_size

    # b. Total Overhead Interrupt (dalam ms)
    total_int_ov_us = num_blocks * int_overhead
    total_int_ov_ms = total_int_ov_us / 1000

    # c. Total Overhead DMA (dalam us) -> sudah diketahui di soal

    # d. Efisiensi
    efficiency = total_int_ov_us / dma_overhead_total

    print(f"a. Jumlah Block: {int(num_blocks)}")
    print(f"b. Overhead Interrupt: {total_int_ov_ms:.2f} ms")
    print(f"c. Overhead DMA: {dma_overhead_total} us")
    print(f"d. DMA {efficiency:.0f}x lebih efisien")

# ==============================================================================
# SOAL 9: Disk Throughput (C3) [cite: 559-568]
# ==============================================================================
def solve_soal_9():
    print("\n--- SOAL 9: Disk Throughput ---")
    transfer_rate = 200 # MB/s
    seek_time = 6       # ms
    rot_latency = 3     # ms
    num_files = 50
    file_size_kb = 8
    file_size_mb = file_size_kb / 1024

    # a. Waktu baca 1 file
    # Transfer time = Size / Rate
    transfer_time_sec = file_size_mb / transfer_rate
    transfer_time_ms = transfer_time_sec * 1000

    total_time_per_file = seek_time + rot_latency + transfer_time_ms

    # b. Total waktu 50 file
    total_time_all_sec = (total_time_per_file * num_files) / 1000

    # c. Throughput Efektif (Total Size / Total Time)
    total_size_mb = num_files * file_size_mb
    throughput = total_size_mb / total_time_all_sec

    print(f"a. Waktu per file: {total_time_per_file:.4f} ms")
    print(f"b. Total waktu (50 file): {total_time_all_sec:.4f} detik")
    print(f"c. Throughput Efektif: {throughput:.2f} MB/s")

# ==============================================================================
# SOAL 10: Buffer Size (C3) [cite: 572-581]
# ==============================================================================
def solve_soal_10():
    print("\n--- SOAL 10: Buffer Calculation ---")
    video_bitrate = 5 # Mbps
    buffer_duration = 10 # detik

    # a. Data rate (MB/s) -> 1 Byte = 8 bits
    rate_mbs = video_bitrate / 8

    # b. Buffer Size (Rate x Duration)
    buffer_size = rate_mbs * buffer_duration

    # c. Bandwidth drop to 4 Mbps (Consumer needs 5 Mbps)
    # Net flow = Input - Output = 4 - 5 = -1 Mbps
    current_bw = 4 # Mbps
    net_drain_mbps = (video_bitrate - current_bw) / 8 # MB/s drain
    time_to_empty = buffer_size / net_drain_mbps

    print(f"a. Data Rate: {rate_mbs} MB/s")
    print(f"b. Buffer Size: {buffer_size} MB")
    print(f"c. Waktu Buffer Habis: {time_to_empty:.1f} detik")

# EXECUTE ALL
solve_soal_6()
solve_soal_7()
solve_soal_8()
solve_soal_9()
solve_soal_10()
```
