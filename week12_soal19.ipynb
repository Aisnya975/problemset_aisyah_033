```python
# SOAL 19: Smart Buffer Manager Pseudocode

CLASS SmartBuffer:
    min_buffer_ms = 50
    max_buffer_ms = 500
    current_buffer_ms = 200  # Start awal
    packet_queue = Queue()

    FUNCTION OnPacketReceived(packet):
        packet_queue.push(packet)
        AnalyzeNetworkCondition(packet)

    FUNCTION AnalyzeNetworkCondition(packet):
        jitter = CalculateJitter(packet.timestamp)
        packet_loss = CheckSequenceGap(packet.seq_num)

        # ADAPTIVE LOGIC:
        # Jika jaringan buruk (jitter tinggi), perbesar buffer (aman tapi latency naik)
        IF jitter > THRESHOLD_HIGH OR packet_loss > 0:
            current_buffer_ms = MIN(current_buffer_ms + 50, max_buffer_ms)

        # Jika jaringan stabil, perkecil buffer (latency makin real-time)
        ELSE IF jitter < THRESHOLD_LOW:
            current_buffer_ms = MAX(current_buffer_ms - 20, min_buffer_ms)

    FUNCTION GetFrameToPlay():
        # Playback logic
        buffered_time = packet_queue.total_duration()

        IF buffered_time >= current_buffer_ms:
            return packet_queue.pop()
        ELSE:
            # Buffer Underrun handling
            return "WAITING / STRETCH AUDIO"
```
