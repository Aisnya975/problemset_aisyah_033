```python
import math

print("=== SOLUSI PRAKTIKUM FILE SYSTEM (MINGGU 11) ===\n")

# =======================================================
# SOAL 7: KONVERSI UNIX PERMISSION (C3)
# Sumber: [cite: 163-168]
# =======================================================
def solve_soal_7():
    print("--- SOAL 7: Unix Permissions ---")

    def symbolic_to_octal(perm_str):
        mapping = {'r': 4, 'w': 2, 'x': 1, '-': 0}
        if len(perm_str) != 9: return "Invalid Length"

        user = mapping[perm_str[0]] + mapping[perm_str[1]] + mapping[perm_str[2]]
        group = mapping[perm_str[3]] + mapping[perm_str[4]] + mapping[perm_str[5]]
        others = mapping[perm_str[6]] + mapping[perm_str[7]] + mapping[perm_str[8]]
        return f"{user}{group}{others}"

    def octal_to_symbolic(oct_str):
        mapping = {
            '0': '---', '1': '--x', '2': '-w-', '3': '-wx',
            '4': 'r--', '5': 'r-x', '6': 'rw-', '7': 'rwx'
        }
        return "".join([mapping[d] for d in oct_str])

    # Jawaban Soal
    print(f"a. rwxr-x---  -> {symbolic_to_octal('rwxr-x---')}")
    print(f"b. rw-rw-r--  -> {symbolic_to_octal('rw-rw-r--')}")
    print(f"c. 644        -> {octal_to_symbolic('644')}")
    print(f"d. 755        -> {octal_to_symbolic('755')}")

    # Analisis Keamanan (Soal 7 akhir) [cite: 172]
    # File: rw-r----- (640) -> User: RW, Group: R, Others: No Access
    print("e. Analisis 'data.txt' (rw-r-----):")
    print("   - Owner: Bisa Baca & Tulis")
    print("   - Group: Hanya Bisa Baca")
    print("   - Others: Tidak ada akses sama sekali")

# =======================================================
# SOAL 6 & 12: BITMAP & DISK LAYOUT SIMULATOR (C3/C4)
# Sumber: [cite: 152-161, 238-248]
# =======================================================
class DiskSimulator:
    def __init__(self, total_blocks, name):
        self.total_blocks = total_blocks
        self.disk = ['0'] * total_blocks # 0 = Free, 1 = Allocated
        self.name = name

    def allocate(self, start, end, label): # end inclusive based on problem
        for i in range(start, end + 1):
            if i < self.total_blocks:
                self.disk[i] = '1' # Mark as used

    def set_bitmap_from_string(self, bitmap_str):
        # Membersihkan spasi dari string soal
        clean_str = bitmap_str.replace(" ", "")
        self.disk = list(clean_str)
        self.total_blocks = len(self.disk)

    def show_stats(self):
        print(f"\nAnalisis Disk {self.name}:")
        used = self.disk.count('1')
        free = self.disk.count('0')
        print(f"Total Blok: {self.total_blocks}")
        print(f"Terisi (1): {used}")
        print(f"Kosong (0): {free}")

        # Cari Contiguous Free Space
        free_spaces = []
        count = 0
        start_index = -1

        for i, val in enumerate(self.disk):
            if val == '0':
                if count == 0: start_index = i
                count += 1
            else:
                if count > 0:
                    free_spaces.append((start_index, count))
                count = 0
        if count > 0: free_spaces.append((start_index, count)) # Catch last segment

        print("Contiguous Free Spaces (Start Index, Length):")
        print(free_spaces)
        return free_spaces

def solve_soal_6_and_12():
    print("\n--- SOAL 6: Disk Layout Analysis ---")
    # Setup Soal 6 [cite: 152-156]
    # Total 32 blok. A:0-4, B:8-11, C:15-19, D:25-28
    d1 = DiskSimulator(32, "Soal 6")
    d1.allocate(0, 4, "A")
    d1.allocate(8, 11, "B")
    d1.allocate(15, 19, "C")
    d1.allocate(25, 28, "D")

    free_spaces = d1.show_stats()

    # Soal 6c: Alokasi file E (6 blok) [cite: 160]
    max_hole = max([fs[1] for fs in free_spaces]) if free_spaces else 0
    print(f"Bisa alokasi file E (6 blok)? {'YA' if max_hole >= 6 else 'TIDAK'} (Max contiguous: {max_hole})")

    print("\n--- SOAL 12: Bitmap Analysis ---")
    # Setup Soal 12 [cite: 238]
    bitmap_str = "11010110 01101011 11010010 00111100"
    d2 = DiskSimulator(32, "Soal 12")
    d2.set_bitmap_from_string(bitmap_str)
    d2.show_stats()

# EXECUTE
solve_soal_7()
solve_soal_6_and_12()
```
