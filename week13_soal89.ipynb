```python
import math

print("=== SOLUSI PERHITUNGAN MINGGU 13 (WINDOWS OS) ===\n")

# =======================================================
# SOAL 8: Performance Monitoring Calculation (C3)
# Sumber Data: [cite: 70-75]
# =======================================================
def solve_soal_8():
    print("--- SOAL 8: Memory Calculation ---")
    total_ram_gb = 16.0
    available_ram_gb = 4.5
    committed_ram_gb = 14.2

    # a. Persentase Memory Utilization
    # Rumus: (Used / Total) * 100. Used = Total - Available [cite: 227]
    used_ram_gb = total_ram_gb - available_ram_gb
    utilization_percent = (used_ram_gb / total_ram_gb) * 100

    print(f"a. Used Memory: {used_ram_gb} GB")
    print(f"   Utilization: {utilization_percent:.2f}%")

    # b. Memory used by processes
    # Secara kasar, ini adalah Used Memory dikurangi cache jika cache dianggap 'available'
    # untuk reclaim, tapi di Windows Task Manager 'In Use' biasanya (Total - Available).
    print(f"b. Memory in active use: {used_ram_gb} GB")

    # c. Analisis Memory Pressure
    # Committed bytes (14.2 GB) mendekati limit fisik (16 GB).
    # Utilization > 70-80% biasanya indikasi pressure.
    print(f"c. Memory Pressure? YA.")
    print(f"   Alasan: Utilization {utilization_percent:.1f}% tinggi, dan")
    print(f"   Committed memory ({committed_ram_gb} GB) sangat dekat dengan Total RAM.")

# =======================================================
# SOAL 9: Network Configuration (C3)
# Sumber Data: [cite: 83]
# =======================================================
def solve_soal_9():
    print("\n--- SOAL 9: Network Config ---")
    ip_address = "192.168.10.25"
    subnet_mask = "255.255.255.0" # /24

    # a. Network Address
    # Untuk /24, ambil 3 oktet pertama + .0
    network_addr = "192.168.10.0"

    # b. Broadcast Address
    # Untuk /24, ambil 3 oktet pertama + .255 [cite: 233]
    broadcast_addr = "192.168.10.255"

    # c. Jumlah Host
    # Rumus: 2^(32 - subnet_bits) - 2 [cite: 230]
    # Subnet /24 berarti 32-24 = 8 bit host
    num_hosts = (2 ** 8) - 2

    print(f"a. Network Address: {network_addr}")
    print(f"b. Broadcast Address: {broadcast_addr}")
    print(f"c. Jumlah Host: {num_hosts} hosts")
    print(f"d. Fungsi DNS 8.8.8.8: Google Public DNS resolver untuk menerjemahkan domain ke IP.")

# EKSEKUSI
solve_soal_8()
solve_soal_9()
```
