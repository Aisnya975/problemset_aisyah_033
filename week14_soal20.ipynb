```python
STRUKTUR DATA LogEntry:
    timestamp: Long (Unix Time)
    user_id: String
    action: Enum (READ, WRITE, DELETE, LOGIN_FAIL)
    resource_id: String
    outcome: Enum (SUCCESS, DENIED)
    prev_hash: String (Blockchain-like linking untuk integritas)
    signature: String (HMAC)

CLASS SecureLogger:
    last_entry_hash = "GENESIS_HASH"

    FUNCTION WriteLog(user, action, resource, outcome):
        entry = NEW LogEntry
        entry.timestamp = NOW()
        entry.user = user
        entry.action = action
        entry.prev_hash = this.last_entry_hash  # Chaining

        # Hitung Signature untuk mencegah tampering
        raw_data = Serialize(entry)
        entry.signature = HMAC_SHA256(raw_data, SECRET_KEY)

        # Simpan ke Write-Once-Read-Many (WORM) storage jika ada
        AppendToDisk(entry)

        # Update hash pointer
        this.last_entry_hash = HASH(entry.signature)

    FUNCTION VerifyLogIntegrity():
        # Audit Function
        current_hash = "GENESIS_HASH"
        FOR each entry IN LogDatabase:
            IF entry.prev_hash != current_hash:
                ALERT "Log Tampering Detected at " + entry.timestamp
            current_hash = HASH(entry.signature)
```
