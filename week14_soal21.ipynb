```python
ALGORITMA: ZeroDayDetector
MODEL: IsolationForest atau AutoEncoder (Trained on Normal Behavior)

FUNCTION TrainingPhase(NormalLogs):
    # Fitur: CPU Usage, Syscall Frequency, Network Outbound Bytes, File Access Rate
    features = ExtractFeatures(NormalLogs)
    Model.train(features)
    Threshold = CalculateErrorThreshold(Model)

FUNCTION RealTimeDetection(CurrentProcess):
    # 1. Ekstraksi Fitur Real-time
    vector = ExtractFeatures(CurrentProcess)

    # 2. Hitung Anomaly Score
    score = Model.predict_score(vector)

    # 3. Decision Logic
    IF score > Threshold:
        # ANOMALI TERDETEKSI (Potensi Zero-day)

        # Langkah A: Isolasi Proses (Sandbox)
        PauseProcess(CurrentProcess.PID)

        # Langkah B: Human Verification / Heuristic Check
        IF HeuristicCheck(CurrentProcess) == MALICIOUS:
            KillProcess(CurrentProcess.PID)
            AlertAdmin("Zero-Day Blocked")
            UpdateModel(vector, label="ATTACK") # Active Learning
        ELSE:
            ResumeProcess(CurrentProcess.PID)
            UpdateModel(vector, label="NORMAL") # Reduce False Positive
```
