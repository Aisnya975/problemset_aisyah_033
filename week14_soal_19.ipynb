```python
ALGORITMA: Lightweight_IoT_MFA
INPUT: UserID, SensorData, Timestamp
OUTPUT: Access_Granted / Denied

CONST SHARED_KEY = "PreSharedKey_LowPower"

FUNCTION Authenticate(DeviceID, UserInput):
    # Faktor 1: "Something You Have" (Device Hardware ID)
    # Cek MAC Address atau Serial Number yang terdaftar
    IF NOT IsWhitelisted(DeviceID):
        RETURN Denied

    # Faktor 2: "Something You Know" (Challenge-Response Ringan)
    # Server kirim random NONCE (Number used once)
    server_nonce = ReceiveNonceFromServer()

    # Device menghitung Hash ringan (misal: Blake2s atau SHA-256)
    # Hash = HASH(SharedKey + server_nonce)
    device_response = CalculateHash(SHARED_KEY, server_nonce)

    IF VerifyWithServer(device_response) == FAIL:
        RETURN Denied

    # Faktor 3: "Context/Behavior" (Something You Do/Are)
    # Menggunakan sensor data sebagai verifikasi lokasi/pola
    # Contoh: Cek apakah GPS device sesuai dengan lokasi rumah user
    current_gps = GetGPSData()
    IF NOT IsSafeLocation(current_gps):
        # Jika lokasi aneh, minta input tambahan (biometrik/PIN)
        RETURN RequestStepUpAuth()

    RETURN Access_Granted
```
