```python
%%writefile structs_example.cpp
#include <iostream>
#include <cstdint>

// Desain Metadata untuk IoT Sensor File System (Soal 18)
// Hemat memori (Packed Struct)

// It's good practice to use 'pragma pack' or similar if you intend
// for structs to be 'packed' as the original comment suggested,
// but for a simple compilation demo, standard struct definitions suffice.
struct BlockHeader {
    uint32_t timestamp;    // Waktu pencatatan (4 bytes)
    uint16_t sensor_id;    // ID Sensor (2 bytes)
    uint16_t data_len;     // Panjang data valid (2 bytes)
    uint32_t next_block;   // Pointer ke blok selanjutnya (Linked Allocation)
};
// Total Header Overhead: 12 bytes per 512 bytes block.
// Efisiensi Storage: (500 / 512) * 100% = ~97.6%

struct DirectoryEntry {
    char sensor_name[8];   // Nama pendek sensor
    uint32_t start_block;  // Blok awal
    uint32_t end_block;    // Blok akhir (untuk append cepat O(1))
};

int main() {
    // Create instances of the structs to show they can be used
    BlockHeader header;
    header.timestamp = 1678886400; // Example timestamp
    header.sensor_id = 101;
    header.data_len = 500;
    header.next_block = 0;

    DirectoryEntry entry;
    // Using snprintf for safe string copying
    snprintf(entry.sensor_name, sizeof(entry.sensor_name), "Temp001");
    entry.start_block = 1;
    entry.end_block = 5;

    std::cout << "BlockHeader Size: " << sizeof(BlockHeader) << " bytes" << std::endl;
    std::cout << "DirectoryEntry Size: " << sizeof(DirectoryEntry) << " bytes" << std::endl;
    std::cout << "Header Timestamp: " << header.timestamp << std::endl;
    std::cout << "Entry Sensor Name: " << entry.sensor_name << std::endl;

    return 0;
}
```
