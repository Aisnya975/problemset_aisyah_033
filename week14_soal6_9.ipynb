```python
import math

print("=== SOLUSI PRAKTIKUM KEAMANAN SISTEM (MINGGU 14) ===\n")

# ==============================================================================
# SOAL 6 & 9: Unix Permission Parser (C3)
# Sumber: [cite: 1537-1540], [cite: 1554-1556]
# ==============================================================================
def parse_permissions(perm_str, filename, user_running="student"):
    print(f"--- Analisis File: {filename} ({perm_str}) ---")

    # Mapping permission
    types = {'-': 'File', 'd': 'Directory', 'l': 'Link'}
    file_type = types.get(perm_str[0], 'Unknown')

    # Owner, Group, Others
    owner_perm = perm_str[1:4]
    group_perm = perm_str[4:7]
    other_perm = perm_str[7:10]

    print(f"a. Penjelasan:")
    print(f"   - Tipe: {file_type}")
    print(f"   - Owner ({owner_perm}): Read, Write, {'SETUID' if 's' in owner_perm else 'Execute'}")
    print(f"   - Group ({group_perm}): Read, {'SETGID' if 's' in group_perm else 'Execute'}")
    print(f"   - Others ({other_perm}): Execute only")

    # Analisis Special Bit
    if 's' in owner_perm:
        print(f"b. Privilege Analysis (SetUID):")
        print(f"   Script berjalan dengan hak akses OWNER (root/admin), bukan user '{user_running}'.")
    elif 's' in group_perm:
        print(f"b. Privilege Analysis (SetGID):")
        print(f"   Proses berjalan dengan hak akses GROUP pemilik file.")
    else:
        print(f"b. Privilege Analysis:")
        print(f"   Proses berjalan dengan hak akses user '{user_running}'.")
    print("-" * 40)

# ==============================================================================
# SOAL 7: Password Cracking Calculation (C3)
# Sumber: [cite: 1545-1547], [cite: 1666-1668]
# ==============================================================================
def solve_soal_7():
    print("\n--- SOAL 7: Perhitungan Brute Force ---")

    # Kasus A: 6 huruf kecil (a-z)
    n_a = 26
    k_a = 6
    speed_a = 1_000_000 # 1 juta/detik
    total_possibility_a = n_a ** k_a
    time_a = total_possibility_a / speed_a

    # Kasus B: 8 karakter (a-z, A-Z, 0-9) -> 26 + 26 + 10 = 62
    n_b = 62
    k_b = 8
    speed_b = 10_000_000 # 10 juta/detik
    total_possibility_b = n_b ** k_b
    time_b = total_possibility_b / speed_b

    print(f"a. 6 huruf kecil (Speed 1M/s):")
    print(f"   Total Kombinasi: {total_possibility_a:,}")
    print(f"   Waktu Maksimal : {time_a:.2f} detik (~{time_a/60:.2f} menit)")

    print(f"b. 8 karakter Alphanumeric (Speed 10M/s):")
    print(f"   Total Kombinasi: {total_possibility_b:,}")
    print(f"   Waktu Maksimal : {time_b:.2f} detik")
    print(f"   Konversi Waktu : {time_b/3600:.2f} jam ({time_b/(3600*24):.2f} hari)")

# ==============================================================================
# SOAL 8: ACL Representation (C3)
# Sumber: [cite: 1549-1552]
# ==============================================================================
def solve_soal_8():
    print("\n--- SOAL 8: Representasi Access Control List (ACL) ---")
    # Struktur Data ACL biasanya berbasis Objek (File)

    acl_system = {
        "FileA": [("Alice", "RWD"), ("Bob", "R"), ("Carol", "RW")],
        "FileB": [("Alice", "R"), ("Bob", "RW"), ("Carol", "RWD")],
        "FileC": [("Alice", "RW"), ("Bob", "RWD"), ("Carol", "R")]
    }

    import json
    print("Representasi JSON (Data Structure):")
    print(json.dumps(acl_system, indent=2))

# ==============================================================================
# SOAL 10: ACL vs Capability List Storage (C3)
# Sumber: [cite: 1558-1560]
# ==============================================================================
def solve_soal_10():
    print("\n--- SOAL 10: Storage Calculation ---")
    users = 200
    files = 50

    # Asumsi: Access Control Matrix jarang penuh (Sparse).
    # Namun soal biasanya meminta worst-case atau explicit entries.
    # Jika setiap user punya akses ke setiap file:
    total_entries = users * files

    print(f"Jumlah User: {users}, Jumlah File: {files}")

    # a. ACL (Disimpan per File)
    # File menyimpan list user yang punya akses.
    print(f"a. Access Control List (ACL):")
    print(f"   Disimpan di 50 File header.")
    print(f"   Jika matriks penuh: {files} list x {users} entry = {total_entries} entri total.")

    # b. Capability List (Disimpan per User)
    # User menyimpan list tiket/kunci ke file.
    print(f"b. Capability List (C-List):")
    print(f"   Disimpan di 200 User profile/process.")
    print(f"   Jika matriks penuh: {users} list x {files} entry = {total_entries} entri total.")
    print("   Catatan: Jumlah total entri sama, bedanya adalah dimana data disimpan.")

# EKSEKUSI
parse_permissions("-rwsr-x--x", "backup.sh", "student") # Soal 6
parse_permissions("-rwxr-sr-x", "project_file", "student") # Soal 9
solve_soal_7()
solve_soal_8()
solve_soal_10()
```
