```python
# Buat folder dulu
New-Item -Path "C:\Projects\Confidential" -ItemType Directory -Force

# Matikan inheritance (Copy existing permissions first -> 'r' removes them, 'd' converts/copies)
# Disini kita remove inheritance agar clean (sesuai requirement "All other users: No access")
icacls "C:\Projects\Confidential" /inheritance:r

# a. Alice: Full Control (F)
icacls "C:\Projects\Confidential" /grant "Alice:(OI)(CI)F"

# b. Bob: Read & Execute (RX)
icacls "C:\Projects\Confidential" /grant "Bob:(OI)(CI)RX"

# c. Managers: Modify (M)
icacls "C:\Projects\Confidential" /grant "Managers:(OI)(CI)M"

# Keterangan: (OI) = Object Inherit, (CI) = Container Inherit
```
