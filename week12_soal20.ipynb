```python
# SOAL 20 (Option A): Priority-Based I/O Scheduler

QUEUE critical_queue  # Priority High (Heart Rate, O2 Sensor)
QUEUE normal_queue    # Priority Low (Log Data, Temp Sensor)

FUNCTION ScheduleIO():
    WHILE True:
        # 1. Strict Priority: Selalu cek queue critical dulu
        IF NOT critical_queue.isEmpty():
            task = critical_queue.dequeue()

            # Cek Deadline (Hard Real-time)
            IF CurrentTime() + task.exec_time > task.deadline:
                Alert("DEADLINE MISS IMMINENT!")
                force_execute_immediately(task)
            ELSE:
                execute(task)

        # 2. Jika Critical kosong, baru kerjakan Normal
        ELSE IF NOT normal_queue.isEmpty():
            task = normal_queue.dequeue()
            execute(task)

        # 3. Anti-Starvation Mechanism untuk Normal Queue
        # Jika task normal menunggu terlalu lama (> 1 detik), naikkan ke critical
        CheckStarvation(normal_queue)

FUNCTION OnNewRequest(request):
    IF request.type == "LIFE_SUPPORT":
        request.deadline = CurrentTime() + 10ms
        critical_queue.enqueue(request)
    ELSE:
        normal_queue.enqueue(request)
```
